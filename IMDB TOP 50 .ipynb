```python
html_text=requests.get('https://www.imdb.com/chart/top/').text

soup=bs4.BeautifulSoup(html_text,'html')
my_dict={}
for x in range(0,50):## 50 STANDS FOR THE 50 MOVIES
    ## NAME OF THE MOVIE
    name=soup.find_all('td',class_="titleColumn")[x].a.text
    print(name)
    ##RELEASE YEAR OF THE MOVIE
    release_year=soup.find_all('td',class_="titleColumn")[x].span.text.replace('(','').replace(')','')
    ##LINK 
    website='https://www.imdb.com'+soup.find_all('td',class_="titleColumn")[x].a['href']+'?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=1a264172-ae11-42e4-8ef7-7fed1973bb8f&pf_rd_r=RF61YJ25BVN9CTR6TBM7&pf_rd_s=center-1&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_tt_1'
    ## REQUESTING THE NEW LINK
    html_text=requests.get(website).text
    soup_latest=bs4.BeautifulSoup(html_text,'html')
    
    ##LIST OF STARS
    new_list_1=[]
    for x in range(0,len(soup_latest.find_all('ul',class_="ipc-metadata-list ipc-metadata-list--dividers-all title-pc-list ipc-metadata-list--baseAlt")[0].find_all('li',class_="ipc-metadata-list__item")[2].find_all('a',class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link"))):
        star=soup_latest.find_all('ul',class_="ipc-metadata-list ipc-metadata-list--dividers-all title-pc-list ipc-metadata-list--baseAlt")[0].find_all('li',class_="ipc-metadata-list__item")[2].find_all('a',class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link")[x].text
        new_list_1.append(star)
     
    ##LIST OF DIRECTORS
    new_list_2=[]    
    for x in range(0,len(soup_latest.find_all('ul',class_="ipc-metadata-list ipc-metadata-list--dividers-all title-pc-list ipc-metadata-list--baseAlt")[0].find_all('li',class_="ipc-metadata-list__item")[0].find_all('a',class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link"))):
        director=soup_latest.find_all('ul',class_="ipc-metadata-list ipc-metadata-list--dividers-all title-pc-list ipc-metadata-list--baseAlt")[0].find_all('li',class_="ipc-metadata-list__item")[0].find_all('a',class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link")[x].text
        new_list_2.append(director)
        
    ##STORING DATA IN A DICTIONARY    
    my_dict[name]={}
    my_dict[name]['RELEASE_YEAR']=release_year
    my_dict[name]['STARS']=new_list_1
    my_dict[name]['DIRECTOR']=new_list_2
        
    
    
```
